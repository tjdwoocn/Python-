### 19.3.1 대각선으로 별 출력하기

```python
for i in range(5):  # 세로
    for j in range(5):  # 가로
        if j==i:    # 세로 방향 변수와 같을때
            print('*', end='')  # 별 출력, 줄바꿈 하지 않음
    print()     # 가로 방향으로 별을 다 그린 뒤 다음 줄로 넘어감
                # print() 는 기본적으로 출력 후 다음 줄로 넘어감
```

- 하지만 별을 출력해보면 그냥 세로로 출력 됨
- IDLE 이나 콘솔에서는 문자 출력을 하지 않으면, 다음 문자는 현재 위치에 출력되므로 모두 왼쪽에 몰림
- 별을 출력하지 않는 부분에는 공백을 출력해주어야 함

```python
for i in range(5):  # 세로
    for j in range(5):  # 가로
        if j==i:    # 세로 방향 변수와 같을때
            print('*', end = '')  # 별 출력, 줄바꿈 하지 않음
        else:
            print(' ', end = '')
    print()     # 가로 방향으로 별을 다 그린 뒤 다음 줄로 넘어감
                # print() 는 기본적으로 출력 후 다음 줄로 넘어감
```

### 20.2 3의 배수일 때와 5의 배수일 때의 처리

```python
for i in range(1, 101):
    if i % 3 == 0:
        print('Fizz')
    if i % 5 == 0:
        print('Buzz')
    else: 
        print(i)
```

### 20.3 3과 5의 공배수 처리

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 ==0:  # 공배수를 먼저 체크 해줘야 함
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else: 
        print(i)
```

### 20.4 논리 연산자를 사용하지 않고 3과 5의 공배수 처리하기

```python
# and 연산자 없이 하려면
# 3 * 5 = 15, 3과 5의 최소공배수인 15를 나눴을때 0인 애들은 3과 5의 공배수임
for i in range(1, 101):
    if i % 15 ==0:  # 공배수를 먼저 체크 해줘야 함
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else: 
        print(i)
```
