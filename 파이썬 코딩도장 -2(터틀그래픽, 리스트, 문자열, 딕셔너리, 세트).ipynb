### 21.2 터틀 그래픽스로 다각형 그리기

```python
import turtle as t
```

```python
n = 6  # 육각형
t.shape('turtle')
t.color('red')
t.begin_fill()  # 색칠할 영역 시작
for i in range(n):
    t.forward(100)
    t.right(360/n)  # 360을 n으로 나누어서 외각을 구함
t.end_fill()   # 색칠할 영영 끝
```

```python
a = []

for i in range(4):
    line = []          # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(5):
        line.append(0) # 안쪽 리스트에 0 추가
    a.append(line)     # 전체 리스트에 안쪽 리스트를 추가
    
print(a)
```

### 톱니형 리스트 만들기

```python
a = [3,1,3,2,5]   # 가로크기 저장
b = []          # 빈 리스트

for i in a:   # 가로 크기를 저장한 리스트 반복
    line =[]   # 안쪽 빈 리스트 생성
    for j in range(i):  # 리스트 a에 저장된 가로 크기만큼 반복
        line.append(0)
    b.append(line)      # 리스트 b에 안쪽 리스트 추가
    
print(b)
```

### 26.2 집합연산

```python
a = {1,2,3,4}
b = {3,4,5,6}
print(a|b)                   # 합집합
print(set.union(a,b))        # 합집합
print(a&b)                   # 교집합
print(set.intersection(a,b)) # 교집합
print(a-b)                   # 차집합 
print(set.difference(a,b))   # 차집합, a 기준
print(set.difference(b,a))   # 차집합, b 기준
print(a^b)                   # 대칭 차집합
print(set.symmetric_difference(a,b))  # 대칭 차집합
```
